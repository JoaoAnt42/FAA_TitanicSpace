```python
import plotly.express as px
import warnings
from lazypredict.Supervised import LazyClassifier
import joblib

warnings.filterwarnings('ignore')
```

```python
X_train, y_train, X_test, y_test = joblib.load('data_FAA/data.pkl')
report = 0
```

```python
clf = LazyClassifier(verbose=0,
                     ignore_warnings=True,
                     custom_metric=None,
                     predictions=False,
                     random_state=12,
                     classifiers='all')

models, predictions = clf.fit(X_train , X_test , y_train , y_test)
```

```python
models
```

```python
if report:
    line = px.line(data_frame= models ,y =["ROC AUC" , "F1 Score"] , markers = True, width=600, height=600)
else:
    line = px.line(data_frame= models ,y =["ROC AUC" , "F1 Score"] , markers = True)

line.update_xaxes(title="Model",
              rangeslider_visible = False)
line.update_yaxes(title = "ROC AUC Score")
line.update_layout(showlegend = True,
    title = {
        'text': 'ROC AUC and F1 Score vs Model',
        'y':0.94,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

line.show()
```

```python
if report:
    line = px.line(data_frame= models ,y =["Time Taken"] , markers = True, width=600, height=600)
else:
    line = px.line(data_frame= models ,y =["Time Taken"] , markers = True)
line.update_xaxes(title="Model",
              rangeslider_visible = False)
line.update_yaxes(title = "Time(s)")
line.update_traces(line_color="purple")
line.update_layout(showlegend = False,
    title = {
        'text': 'TIME TAKEN vs Model',
        'y':0.94,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

line.show()

```
